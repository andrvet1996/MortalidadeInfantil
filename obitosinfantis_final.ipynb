5) APENAS BARUERI (2000-2021)**

```r
obitosbarutotal = filter(obitoslimpos, cidade == 3505708)
obitosbarutotal
```

```r
attach(obitosbarutotal)
summary(obitosbarutotal)
```

```r
sd(obitosbarutotal$vivos)
sd(obitosbarutotal$perinatais)
sd(obitosbarutotal$maformacao)
sd(obitosbarutotal$respiratorio)
sd(obitosbarutotal$infectparasit)
```

```r
boxplot(obitosbarutotal$vivos, ylab = 'Nascidos vivos', ylim = c(4500,6000), main = 'Total de nascidos vivos entre 2000 a 2021')
```

```r
boxplot(obitosbarutotal$perinatais, ylab = 'Total de óbitos perinatais', ylim = c(0,50), main = 'Óbitos perinatais entre 2000 a 2021')
```

```r
boxplot(obitosbarutotal$maformacao, ylab = 'Total de obitos por má formação', ylim = c(0,50), main = 'Óbitos por má formação entre 2000 a 2021')
```

```r
boxplot(obitosbarutotal$respiratorio, ylab = 'Total de óbitos por doença respiratória', ylim = c(0,50), main = 'Óbitos por doença respiratória entre 2000 a 2021')
```

```r
boxplot(obitosbarutotal$infecparasit, ylab = 'Total de óbitos por doença infecciosa/parasitária', ylim = c(0,50), main = 'Óbitos por doença infecciosa/parasitária entre 2000 a 2021')
```

```r
hist(obitosbarutotal$vivos, xlab = 'Total de nascidos vivos', main = 'Total de nascidos vivos entre 2000 e 2021', probability = TRUE)
lines(density(obitosbarutotal$vivos))
```

```r
hist(obitosbarutotal$perinatais, xlab = 'Óbitos perinatais', ylim = c(0, 0.08), main = 'Mortalidade perinatal de 2000 a 2021', probability = TRUE)
lines(density(obitosbarutotal$perinatais))
```

```r
hist(obitosbarutotal$maformacao, xlab = 'Óbitos por má formação', xlim = c(5,25), ylim = c(0, 0.15), main = 'Mortalidade por má formação de 2000 a 2021', probability = TRUE)
lines(density(obitosbarutotal$maformacao))
```

```r
hist(obitosbarutotal$respiratorio, xlab = 'Óbitos por doença respiratória', ylim = c(0, 0.25), main = 'Mortalidade por doença respiratória de 2000 a 2021', probability = TRUE)
lines(density(obitosbarutotal$respiratorio))
```

```r
hist(obitosbarutotal$infecparasit, xlab = 'Óbitos por doença infecciosa/parasitária', ylim = c(0, 0.70), main = 'Mortalidade por doença infecciosa/parasitária de 2000 a 2021', probability = TRUE)
lines(density(obitosbarutotal$infecparasit))
```
